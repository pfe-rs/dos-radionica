```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
# rucno implementirana konvolucija
def conv(x, y):
    c = np.zeros(len(x) + len(y) - 1)
    for i in range(len(x)):
        for j in range(len(y)):
            c[i + j] += x[i] * y[j]
    return c
```

```python
t = np.linspace(0, 10, 1000)

# ulazni signal 
x = 1 * (t > 1) * (t < 2)

# impulsni odziv
h = (t > 3) * (t < 4)

# izlazni signal
y = conv(x, h)

t_y = np.linspace(0, 20, len(y)) # povecavamo opseg vremenske ose da bi se pravilno prikazalo

fig, ax = plt.subplots(3,1)
ax[0].plot(t, x)
ax[0].set_xlabel('t')
ax[0].set_ylabel('x(t)')
ax[0].set_title('Ulazni signal')

ax[1].plot(t, h)
ax[1].set_xlabel('t')
ax[1].set_ylabel('h(t)')
ax[1].set_title('Impulsni odziv')

ax[2].plot(t_y, y)
ax[2].set_xlabel('t')
ax[2].set_ylabel('y(t)')
ax[2].set_title('Izlazni signal')
ax[2].set_xlim(0, 10)
ax[2].set_xticks([0, 2, 4, 6, 8, 10])
fig.set_size_inches(12,16)
```
