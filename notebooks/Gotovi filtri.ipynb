```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy import signal
```

### Primena butterworth filtra

```python
t = np.linspace(0, 1, 1000)  
x = np.sin(2*np.pi*10*t) + np.sin(2*np.pi*20*t) + np.sin(2*np.pi*30*t) # signal sa sinusoidama na 10, 20 i 30 Hz
fig, ax = plt.subplots(2, 1)
ax[0].plot(t, x)
ax[0].set_title('Signal sa sinusoidama na 10, 20 i 30 Hz')
ax[1].axis([0, 1, -2, 2])

sos = signal.butter(10, 25, 'low', fs=1000, output='sos')
filtered = signal.sosfilt(sos, x)
ax[1].plot(t, filtered)
ax[1].set_title('Posle lowpass filtra sa frekvencijom odsecanja 25 Hz')
ax[1].axis([0, 1, -2, 2])
ax[1].set_xlabel('t')
fig.set_size_inches(10,5)
plt.tight_layout()
plt.show()
```
