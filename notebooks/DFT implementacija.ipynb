```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# stopa odabiranja
sr = 100
# interval odabiranja
ts = 1.0/sr
t = np.arange(0,1,ts)

freq = 1.
x = 3*np.sin(2*np.pi*freq*t)

freq = 4
x += np.sin(2*np.pi*freq*t)

freq = 7   
x += 0.5* np.sin(2*np.pi*freq*t)

plt.plot(t, x)
plt.title("Signal")
plt.xlabel("Vreme [s]")
plt.ylabel("Amplituda")
plt.show()
```

```python
def DFT(x):
    """
    Funkcija koja racuna diskretnu Furijeovu transformaciju 1D realnog signala x
    """

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)
    
    return X
```

```python
X = DFT(x)

# racunanje frekvencija
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (8, 6))
plt.stem(freq, abs(X), linefmt="b", markerfmt=" ", basefmt="-b")
plt.title("DFT signala")
plt.xlabel('Frekvencija [Hz]')
plt.ylabel('Amplituda')
plt.show()
```

```python
n_oneside = N//2
# uzmimanje frekvencija za jednu stranu
f_oneside = freq[:n_oneside]

# normalizacija amplitude
X_oneside =X[:n_oneside]/n_oneside

plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.stem(f_oneside, abs(X_oneside), linefmt='b', markerfmt=" ", basefmt="-b")
plt.xlabel('Frekvencija [Hz]')
plt.ylabel('Normalizovana amplituda')

plt.subplot(122)
plt.stem(f_oneside, abs(X_oneside), linefmt='b', markerfmt=" ", basefmt="-b")
plt.xlabel('Frekvencija [Hz]')
plt.xlim(0, 10)
plt.tight_layout()
plt.show()
```
